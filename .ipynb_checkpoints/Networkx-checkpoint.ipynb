```python
# For creating networks 
import networkx
G=networkx.Graph()
```

```python
G.add_node(1)
G.add_node(2)
G.add_node(3)
G.add_node(4)
G.add_node(5)
```

```python
G.nodes()
```

```python
G.add_edge(1,2)
G.add_edge(2,3)
G.add_edge(1,3)
G.add_edge(4,5)
G.add_edge(1,4)
```

```python
G.edges()
```

```python
# For visualization
import matplotlib.pyplot as plt
networkx.draw(G,with_labels=1)
plt.show
```

```python
import networkx as nx
Z=nx.gnp_random_graph(10,0.5)
nx.draw(Z,with_label=1)
plt.show
```
